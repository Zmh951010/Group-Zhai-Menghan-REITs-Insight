```python
from pyecharts import Line
```

```python
line = Line("Different Investment Return Comparison (US, monthly, 2008-2018) ",width=1000, height=600)
line.add("REITs Return", df['time'], df['REITsReturn'], mark_line=["average"])
line.add("S&P500", df["time"], df['S&P500'], mark_line=["average"])
line.add("NASDAQ",df["time"], df['NASDAQComposite'], mark_line=["average"])
line.add("Small Cap Stocks", df["time"], df['Russell2000'], mark_line=["average"])
line.add("Investment Bond", df["time"], df['Bond'], mark_line=["average"])
line.add("House Sales", df["time"], df['NewHouseSalesGrowth'], mark_line=["average"])
line.add("10-year bond yield", df["time"], df['TreasuryBondYield'], mark_line=["average"])
line.add("House Price", df["time"], df['FHFAHousePriceIndex'], mark_line=["average"])
line.add("inflation", df["time"], df['CPIgrowth'], mark_line=["average"])
line.add("interest rate", df["time"], df['FederalFundRate'], mark_line=["average"],
         legend_orient='vertical',
         is_datazoom_show=True,
         datazoom_type='slider',
         legend_top='top',
         legend_pos='right'
        
        )
line.render('UScomparison.html')
```

```python
tenyears=list(df.loc[:,'REITsReturn':'Bond'].sum())
```

```python
fiveyears=list(df.loc[68:128, 'REITsReturn':'Bond'].sum())
```

```python
oneyear=list(df.loc[116:128, 'REITsReturn':'Bond'].sum())
```

```python
attr=['REITs Return', 'S&P500', 'NASDAQ', 'Small Cap Stocks','Investment Bond']
```

```python
from pyecharts import Bar
```

```python
bar = Bar("10-Year/5-Year/1-Year Accumulated Return Rate Comparison (US, 2008-2018)", width=1200, height=600)
bar.add("10-Year", attr, tenyears)
bar.add("5-Year", attr, fiveyears)
bar.add("1-Year", attr, oneyear,
        legend_orient='vertical',
        
        legend_top='10%',
        legend_pos='right')
bar.render('US investments accumulated return.html')
```

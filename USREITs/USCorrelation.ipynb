```python
import pandas as pd
```

```python
df = pd.read_csv('US1.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
corrmat=df.corr()
```

```python
corrmat
```

```python
x_axis = ['REITsReturn', 'S&P500', 'NASDAQ', 'SmallCap', 'Bond',
       'HouseSales', 'HousePrice', 'TreasuryBond',
       'FFR', 'CPI']
y_axis = ['REITsReturn', 'S&P500', 'NASDAQ', 'SmallCap', 'Bond',
       'HouseSales', 'HousePrice', 'TreasuryBond',
       'FFR', 'CPI']
```

```python
data = [[i,j,corrmat.iloc[i,j]] for i in range(0,10) for j in range (0,10)]
```

```python
data
```

```python
from pyecharts import HeatMap
#data = [[i, j, random.randint(0, 50)] for i in range(24) for j in range(7)]
heatmap = HeatMap('Correlation Matrix of Different Type of Investment', width=1200, height=1050)
heatmap.add("Correlation among products",
    x_axis,
    y_axis,
    data,
    xaxis_interval=0,
    xaxis_pos='top',
    yaxis_interval=0,
    is_yaxis_inverse=True,
    is_label_show=True,
    is_visualmap=True,
    visual_text_color="#000",
    visual_range_color=['#1E90FF','#FFFFFF','#FF0000'],
    visual_range=[-1, 1],
    visual_pos='right',
    visual_top='top',
   is_legend_show=False
           )
heatmap.render("UScorr.html")
```
